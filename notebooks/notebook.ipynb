```python
# יצירת מילון המרות אוטומטי מתוך mushroom_dict
value_mappings = {}
for _, row in mushroom_dict.iterrows():
    name = row['name']
    description = row['description']

    if pd.isna(description):
        continue

    # פירוק התיאור לזוגות של ערכים
    mapping = {}
    pairs = description.split(',')
    for pair in pairs:
        if '=' in pair:
            full_val, code = pair.strip().split('=')
            mapping[code] = full_val

    if mapping:
        value_mappings[name] = mapping
value_mappings['poisonous'] = {'p': 'poisonous', 'e': 'edible'}
```
